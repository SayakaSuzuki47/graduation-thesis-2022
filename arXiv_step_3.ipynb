```python
#関係するツールの導入
import os
import numpy as np
import pandas as pd
import collections
import re
from string import digits
```

```python
#read_file_names_list (str)-> str_lst
#("入力の場所(フォルダまでのパス)")->"入力の場所"にあるファイル名を入れたstring型のlistを返す
#
#"入力の場所"はあるフォルダへのパス.
#フォルダの中にあるファイルの名前を読み込む.
#ファイルの名前をリストに入れていく.
#ファイルの名前を入れたリストを出力として返す.
#
#例
#"/file"->["folder_1.txt","folder_2.txt","folder_3.txt"]
#
# read_file_names_list(input_path)
#
#param string input_path 読み込みファイルが存在するフォルダパス
#
#return string[] 分割されたファイル名のリスト
#

def read_file_names_list(input_path):
  files_list = []
  for filename in os.listdir(input_path):
    if os.path.isfile(os.path.join(input_path, filename)): #ファイルのみ取得
      files_list.append(filename)
  return(files_list)
```

```python
#read_lst (str,str)->dct
#(入力の場所,ファイルの名前)->単語を数えてdct型にする(key,value)->(str,int)
def read_lst (input_folder_path, output_folder_path, file_name):
  file_path = input_folder_path + "/" + file_name
  f = open(file_path,"r")
  list_row = []
  for x in f:
    list_row.append(x.rstrip("\n"))
    #以下のようにしてしまうと、改行コードがlistに入ってしまうため注意
    #list_row.append(x)
  f.close()
  word_lst = collections.Counter(list_row)
  df = pd.DataFrame({})
  df = df.append(word_lst, ignore_index=True)
  df = df.fillna(0)
  df = df.astype(int)
  target = '.txt'#ext_target
  idx = file_name.find(target)
  r = file_name[:idx]
  df.to_csv(output_folder_path + '/' + r + '.csv', sep=",")
  os.remove(file_path)
  #os.remove(file_path)
```

```python
#read_lstをlstに対して繰り返し実行する.
txt_data =list(map( lambda x : read_lst(input_folder_path, output_folder_path, x), file_names_list))
```

```python
#read_file_names_list (str,str)-> str_lst
#("入力の場所(フォルダまでのパス)")->"入力の場所"にあるファイル名(拡張子は除く)を入れたstring型のlistを返す
#
#"入力の場所"はあるフォルダへのパス.
#フォルダの中にあるファイルの名前を読み込む.
#ファイルの名前をリストに入れていく.
#ファイルの名前を入れたリストを出力として返す.
#
#例
#"/file"->["folder_1","folder_2","folder_3"]
#
# read_file_names_list(input_path)
#
#param string input_path 読み込みファイルが存在するフォルダパス
#
#return string[] 分割されたファイル名のリスト
#

def read_file_names_list_sub(input_path, file_target):
  files_list = []
  for filename in os.listdir(input_path):
    if os.path.isfile(os.path.join(input_path, filename)): #ファイルのみ取得
      target =  file_target #text_target
      idx = filename.find(target)
      r = filename[:idx]
      files_list.append(r)
  return(files_list)
```
