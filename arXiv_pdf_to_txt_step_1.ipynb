```python
#files_list: path_1にあるファイル名の一覧を獲得する.
#read_file_names_list
file_names_list = []
for filename in os.listdir(input_folder_path):
  if os.path.isfile(os.path.join(input_folder_path, filename)): #ファイルのみ取得
    file_names_list.append(filename)
```

```python
def read_file_names_list(input_path, ext):
  return([os.path.splitext(fn)[0] for fn in os.listdir(input_path) if (os.path.isfile(os.path.join(input_path, fn)) & (os.path.splitext(fn)[1] == ext))])
```

```python
#get_txt_from_pdf (str,str,str)->
#(入力の場所,出力の場所,ファイルの名前)->txtファイルを出力の場所に作る. 
def get_txt_from_pdf (import_folder_path,export_folder_path,file_name):
  pdfpath_1 = import_folder_path + "/" + file_name
  pdfpath_2 = export_folder_path + "/" 
  #pdf2txt.pyを用いてpdfファイルをテキスト変換
  lines = !pdf2txt.py { pdfpath_1 }
  txt = '\n'.join(lines)
  target = '.pdf'#ext_target
  idx = file_name.find(target)
  r = file_name[:idx] 
  rename = pdfpath_2 + r + ".txt"
  f= open( rename ,"w")
  f.write(txt)
  f.close()
  os.remove(pdfpath_1)
```

```python
#get_txt_from_pdfをlistに対して繰り返し実行
list(map( lambda x:get_txt_from_pdf(input_folder_path, output_folder_path, x), file_names_list))
```
