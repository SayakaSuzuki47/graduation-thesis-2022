```python
#read_lst(str, str)->dct
#入力のtxtファイルを読み込んでdct型に変換する.
def read_lst (path,file_name):
  file_path = path + "/" + file_name
  f = open(file_path,"r")
  dic = {}
  for line in f:
    # 区切り文字の変更: スペース => カンマ
    line_1 = line.rstrip("\n")
    key, val = line_1.split(", ") 
    dic[key] = int(val)
  f.close()
  return(dic)
```

```python
#dct_freq (dict) -> dict
#dctのvaluの和が1になるようにする
def dct_frec (dct):
  sum_value = sum(dct.values())
  r_dict = { k : v/sum_value for k ,v in dct.items() }
  return(r_dict)
```

```python
#merge_dict_add_values: (dict,dict) -> dict
#dictとdictを足す.keyに対するvalueはd1とd2を足したもの.
#d1は上書きされる.
from collections import Counter

def merge_dict_add_values(d1,d2):
  return dict(Counter(d1) + Counter(d2))
```

```python
#merge_dict_pull_values: dict -> dict
#d2にあるkeyだけでd2.value - d1.value 
def merge_dict_pull_values(d1,d2):
  d3 = {k: (d2.get(k,0) - d1.get(k,0)) for k in d2.keys()}
  return d3
```

```python
#上位n個のkeyを取りだす
# make_top_key_list :(dict_list,int)-> str_list
def make_top_key_list (lst,n) :
  list_dictionary_n = []
  for i in range(n):
    (key,value) = lst[i]
    list_dictionary_n.append(key)
  return (list_dictionary_n)
```

```python
#top_dictionaryを降順にsort
top_dictionary_sorted = sorted(top_dictionary.items(), key=lambda x:x[1], reverse=True)
```

```python
#list_dictionary 
#top_dictionaryからtop n個の単語を取りだしたもの
list_dictionary = []
len_dict_top_40par = int(len(top_dictionary)*0.4)
list_dictionary = make_top_key_list(top_dictionary_sorted,len_dict_top_40par)
```

```python
#上位n個のkeyをtxt_data_listから取りだす
# make_txt_top_n_dictionary : (int, dict) -> str_list
def make_txt_top_n_dictionary (n,txt_data_list):
  data_1 = sorted(txt_data_list.items(), key=lambda x:x[1], reverse=True)
  number = min(n,len(txt_data_list))
  data_2 = make_top_key_list(data_1, number)
  return data_2
```

```python
n = 10
txt_top_n_dictionary = list(map(lambda x:make_txt_top_n_dictionary(n, x),test_data_4))
```

```python
#add_str_lst_set (list,list) -> list
#list1は上書きされる
def add_str_lst_set  (list1,list2):
  #list1とlist2をつなげる
  list1.extend(list2)
  #重複の除外
  list1=list(set(list1))
  return list1
```
