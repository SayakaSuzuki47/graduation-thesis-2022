```python
#read_file_names_list (str)-> str_lst
#("入力の場所(フォルダまでのパス)")->"入力の場所"にあるファイル名を入れたstring型のlistを返す
#
#"入力の場所"はあるフォルダへのパス.
#フォルダの中にあるファイルの名前を読み込む.
#ファイルの名前をリストに入れていく.
#ファイルの名前を入れたリストを出力として返す.
#
#例
#"/file"->["folder_1.txt","folder_2.txt","folder_3.txt"]
#
# read_file_names_list(input_path)
#
#param string input_path 読み込みファイルが存在するフォルダパス
#
#return string[] 分割されたファイル名のリスト
#

def read_file_names_list(input_path):
  files_list = []
  for filename in os.listdir(input_path):
    if os.path.isfile(os.path.join(input_path, filename)): #ファイルのみ取得
      files_list.append(filename)
  return(files_list)
```

```python
file_names_list = read_file_names_list(input_folder_path)
file_names_list.sort(reverse=True)
```

```python
#get_txt_from_pdf (str,str,str)->
#(入力の場所,出力の場所,ファイルの名前)->txtファイルを出力の場所に作る. 
def get_txt_from_pdf (import_folder_path,export_folder_path,file_name):
  pdfpath_1 = import_folder_path + "/" + file_name
  pdfpath_2 = export_folder_path + "/" 
  #pdf2txt.pyを用いてpdfファイルをテキスト変換
  lines = !pdf2txt.py { pdfpath_1 }
  txt = '\n'.join(lines)
  target = '.pdf'#ext_target
  idx = file_name.find(target)
  r = file_name[:idx] 
  rename = pdfpath_2 + r + ".txt"
  f= open( rename ,"w")
  f.write(txt)
  f.close()
```
