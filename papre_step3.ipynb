```python
#files_list: path_1にあるファイル名の一覧を獲得する.
files_list = []
for filename in os.listdir(path_1):
  if os.path.isfile(os.path.join(path_1, filename)): #ファイルのみ取得
    files_list.append(filename)
```

```python
#read_lst (str,str)->dct
#(入力の場所,ファイルの名前)->単語を数えてdct型にする(key,value)->(str,int)
def read_lst (path,file_name):
  file_path = path + "/" + file_name
  def count_words (word_lst):
    words = {}
    for word in word_lst:
    # 単語をキーとして値に1を足していく。
    # 辞書に単語がない、すなわち初めて辞書に登録するときは0+1になる。
      words[word] = words.get(word, 0) +1 
    return words
  f = open(file_path,"r")
  list_row = []
  for x in f:
    list_row.append(x.rstrip("\n"))
    #以下のようにしてしまうと、改行コードがlistに入ってしまうため注意
    #list_row.append(x)
  f.close()
  word_lst = count_words(list_row)
  return(word_lst)
```

```python
#get_txt_from_pdf (str,str,str)->
#(入力の場所,出力の場所,ファイルの名前)->txtファイルを出力の場所に作る. 
def get_dct_of_word (path_a,path_b,file_name):
  ##pdfpath_1 = path_1 + "/" + file_name
  pdfpath_b = path_b + "/" + file_name
  dct_lst = read_lst (path_a,file_name)
  f= open( pdfpath_b ,"w")
  #lst = take_out_word(path_a,file_name)
  for key,value in dct_lst.items():
    f.write('{0}, {1}\n'.format(key, value))
    #f.write(f'{key} \,{value}\n')
  f.close()
```
