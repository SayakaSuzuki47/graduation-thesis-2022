##Bert用にabstractを取り出す

```python
bert_abstract_folder_path =  '/content/drive/MyDrive/Colab_Notebooks/bert/abstract'
```

```python
decoder = json.JSONDecoder()
with open(json_file_path, 'r') as f:
  for i in range(1):
    line = f.readline()
    json_data = decoder.raw_decode(line)
    id = json_data[0]['id']
    abstract = json_data[0]["abstract"]
    f = open(bert_abstract_folder_path + '/' + id + '.txt', 'w')
    f.write(abstract)
    f.close()
```
