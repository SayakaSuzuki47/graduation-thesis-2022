```python
#path_1: 入力となるファイルの保存場所
path_1 = '/content/drive/MyDrive/Colab_Notebooks/paper/pdf_file_test'
#input,import,read _folder_path

#path_2: 出力となるpdfファイルの保存場所
path_2 = '/content/drive/MyDrive/Colab_Notebooks/paper/txt_test'
#output,export,write _folder_path

#files_list: path_1にあるファイル名の一覧を獲得する.
#read_file_names_list
files_list = []
for filename in os.listdir(path_1):
  if os.path.isfile(os.path.join(path_1, filename)): #ファイルのみ取得
    files_list.append(filename)
```

```python
#get_txt_from_pdf (str,str,str)->
#(入力の場所,出力の場所,ファイルの名前)->txtファイルを出力の場所に作る. 
def get_txt_from_pdf (import_folder_path,export_folder_path,file_name):
  pdfpath_1 = import_folder_path + "/" + file_name
  pdfpath_2 = export_folder_path + "/" 
  #pdf2txt.pyを用いてpdfファイルをテキスト変換
  lines = !pdf2txt.py { pdfpath_1 }
  txt = '\n'.join(lines)
  target = '.pdf'#ext_target
  idx = file_name.find(target)
  r = file_name[:idx] 
  rename = pdfpath_2 + r + ".txt"
  f= open( rename ,"w")
  f.write(txt)
  f.close()
```

```python
#get_txt_from_pdfをlistに対して繰り返し実行
list(map( lambda x:get_txt_from_pdf(path_1,path_2,x),files_list))
```
